# Gráfico de pizza a partir de CSV

## Importação de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Leitura do CSV

```python
torcedores = pd.read_csv("torcedores_champions.csv")
```

## Criação do gráfico de pizza

```python
plt.figure(figsize=(7,7))

plt.pie(torcedores["porcentagem"], labels=torcedores["time"], autopct="%1.1f%%", startangle=90)

plt.title("Porcentagem de torcedores - Times da Champions League")
plt.show()
```
